## Data Cleaning
I will mainly be working with WorldBank data regarding the "GDP per capita (current US$)" and the "Gini index (World Bank estimate)" of nations.
\n
At a glance, there are many missing values in the dataset.

```python
#load the gdp_per_capita data
#first four rows of the csv file includes information about last update date and source
gdp_df = pd.read_csv('gdp_per_capita.csv', skiprows = 4)

#drop unnecessary columns/information
gdp_df = gdp_df.drop(columns = ['Indicator Name', 'Indicator Code', 'Unnamed: 65'])

#create new column tracking number of years of missing data per nation
gdp_df['Missing Years'] = gdp_df.isnull().sum(axis=1)
```

```python
#load the gini index data
#first four rows of the csv file includes information about the last update date and source
gini_df = pd.read_csv('gini_index.csv', skiprows = 4)
gini_df.head

#drop unnecessary columns/information
gini_df = gini_df.drop(columns = ['Indicator Name', 'Indicator Code', 'Unnamed: 65'])

#creae new column tracking number of years of missing data per nation
gini_df['Missing Years'] = gini_df.isnull().sum(axis=1)

gini_df
```
